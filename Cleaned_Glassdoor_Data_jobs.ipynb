```python
# read in the csv file as a pandas dataframe
df = pd.read_csv("Uncleaned_DS_jobs.csv")
```

```python
# explore the data
df.info()
```

```python
#Dropping columns that will not be used in this project
df = df.drop(['Job Description', 'Location', 'Headquarters', 'Founded', 'Competitors'], axis = 1)
```

```python
# explore new data
df.info()
```

```python
#Updating the Header name
df = df.rename(columns={'index':'Index', 'Rating':'Company Rating', 'Size':'Company Size', 'Type of ownership':'Type', 'Salary Estimate':'Salary Estimate (In USD)', 'Revenue':'Revenue (in USD)'})
df
```

```python
# Replacing all '-1' values of the dataset with NULL
df['Company Rating'] = df['Company Rating'].astype('float')
df= df.replace(['-1', -1.0] ,'')
df
```

```python
#Cleaning Job Titles
#All Data Engineering job titles renamed as 'Data Engineer'
df.loc[df['Job Title'].str.contains('Engineer', case=False), 'Job Title'] = 'Data Engineer'

#All Data Scientist job titles renamed as 'Data Scientist'
df.loc[df['Job Title'].str.contains('Scientist', case=False), 'Job Title'] = 'Data Scientist'

#All Data Analyst job titles renamed as 'Data Analyst'
df.loc[df['Job Title'].str.contains('Analyst', case=False), 'Job Title'] = 'Data Analyst'

#All other job titles renamed as 'other'
df.loc[~df['Job Title'].str.contains('Data Engineer|Data Scientist|Data Analyst', case=False), 'Job Title'] = 'Other'
```

```python
#Explore the jobs with new titles
df['Job Title'].value_counts()
```

```python
#Extracting MAX salary 

df['Salary Estimate (In USD)'] = df['Salary Estimate (In USD)'].str.split(pat = '-').str[1]
df['Salary Estimate (In USD)'] = df['Salary Estimate (In USD)'].str.split(pat = 'K').str[0]

# Removing 'K' and getting the salary in thousands
df['Salary Estimate (In USD)'] = df['Salary Estimate (In USD)'].str.split(pat = '$').str[1] + '000'

#Converting the salary field to data type 'int'
df['Salary Estimate (In USD)'].astype('int')
```

```python
df['Company Name'].value_counts()
```

```python
#Cleaning Company Size
#All Data Engineering job titles renamed as 'Data Engineer'
df['Company Size'] = df['Job Title'].str.replace('Analyst', 'computer', case=False)
df.info()
```

```python
# Updating 'Company Name'
df['Company Name'] = df['Company Name'].str.split(pat = '\n').str[0]
df
```

```python
#Updating Company Size
df['Company Size'] = df['Company Size'].replace(['-1','Unknown'],'Very Large')
df['Company Size'] = df['Company Size'].replace(['5001 to 10000 employees', '10000+ employees'] ,'Large')
df['Company Size'] = df['Company Size'].replace('1001 to 5000 employees','Medium')
df['Company Size'] = df['Company Size'].replace(['-1','1 to 50 employees', '51 to 200 employees', '201 to 500 employees','501 to 1000 employees'] ,'Small')
df
```

```python
#Replacing unknown revenue values with NULL
df['Revenue (in USD)'] = df['Revenue (in USD)'].replace(['-1', 'Unknown / Non-Applicable'] ,'')
df['Revenue (in USD)']
```

```python

# writing the final dataframe to an excel file that we will use in our Tableau visualizations.
df.to_excel('Cleaned_Glassdoor_Job_Data.xlsx', sheet_name = 'Job_Data')
```
